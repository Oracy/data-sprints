```python
import pandas as pd
from shutil import copyfile
import numpy as np
%matplotlib inline
```

```python
def is_json(json_file):
    try:
        json_object = pd.read_json(json_file)
        json_object = None
        return True
    except ValueError as e:
        return False


def manipulate_wrong_json_list(file_source_list, file_target_list):
    for (idx_file, file_source) in enumerate(file_source_list):
        file_is_json = is_json(file_source)
        if file_is_json != True:
            data = []
            with open(file_source) as fp:
                line = fp.readline()
                cnt = 1

                while line:
                    row = line.strip()
                    data.append(row)
                    line = fp.readline()
                    cnt += 1

            f = open(file_target_list[idx_file], 'w')
            for (idx, i) in enumerate(data):
                if idx == 0:
                    f.write("[{},".format(i))
                elif idx < len(data)-1:
                    f.write("{},".format(i))
                else:
                    f.write("{}]".format(i))
            f.close()
        else:
            copyfile(file_source, file_target_list[idx_file])
```

# Bonus

```python
trip_1["weekday"] = pd.to_datetime(trip_1.pickup_date).dt.dayofweek
```

```python
trip_1['pickup_datetime'] = pd.to_datetime(trip_1['pickup_datetime'])
trip_1['dropoff_datetime'] = pd.to_datetime(trip_1['dropoff_datetime'])
trip_1
```

```python
trip_1['diff_trip_seconds'] = trip_1['dropoff_datetime'] - trip_1['pickup_datetime']
trip_1['diff_trip_seconds'] = trip_1['diff_trip_seconds']/np.timedelta64(1, 's')
trip_1
```

```python
weekend_trips = trip_1[trip_1['weekday'] >=5]
weekend_trips
```

```python
weekend_mean_trip = weekend_trips.groupby(['weekday']).agg({'diff_trip_seconds': "mean"})
weekend_mean_trip
```
